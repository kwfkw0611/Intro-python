### Convert datetime datatype
`.to_datetime()` to convert a datetime argument to datetime data object

```python
df_aapl['Date'] = pd.to_datetime(df_aapl['Date']) 
```

### Resample
`.resample()` method for frequency conversion and resampling of time series, object must have datetime-like index 

Table for common time series frequencies, pass in alias as a string

Alias | Description
---- | ----
B | business day frequency
C | custom business day frequency
D | calendar day frequency
W | weekly frequency
M | month end frequency
SM | semi-month end frequency (15th and end of month)
BM | business month end frequency
CBM | custom business month end frequency
MS | month start frequency
SMS | semi-month start frequency (1st and 15th)
BMS | business month start frequency
CBMS | custom business month start frequency
Q | quarter end frequency
BQ | business quarter end frequency
QS | quarter start frequency
BQS | business quarter start frequency
A, Y | year end frequency
BA, BY | business year end frequency
AS, YS | year start frequency
BAS, BYS | business year start frequency
BH | business hour frequency
H | hourly frequency
T, min | minutely frequency
S | secondly frequency
L, ms | milliseconds
U, us | microseconds
N | nanoseconds

```python
df_aapl.resample('M') # returns a resampler object
```

```python
df_aapl.resample('M').mean() # to get output result need an aggregate method
```

```python
# alternative syntax using .apply()
df_aapl.resample(rule='M').apply('mean') # pass a built-in python function as string
```

```python
# with .apply() can also pass in user defined function
def avg(col): 
    a = sum(col) / len(col) 
    return a

df_aapl.resample(rule='M').apply(avg)
```

### Pandas plotting on resampled data


```python
# default plot is line
df_aapl['Close'].resample('M').mean().plot() 
```

```python
# to change plot style 
df_aapl['Close'].resample('M').mean().plot(kind='bar',figsize=(12,6))
```

### Pivot table

[Reshaping and pivot tables pandas docs](https://pandas.pydata.org/pandas-docs/stable/reshaping.html#reshaping)

```python
df_aapl.head()
```

```python
# default aggregate function is mean
df_aapl.pivot_table(index='month', values=['Open', 'High', 'Low', 'Close'])
```

```python
# can pass in multiple aggregate functions
df_aapl.pivot_table(index='month', values=['Open', 'High', 'Low', 'Close'], aggfunc=['mean','median', 'max', 'min', 'std'])
```

```python
# plotting pivot table
pd.pivot_table(df_aapl, index='month', values=['Close', 'High', 'Low', 'Open']).plot(kind='bar',figsize=(12,6))
```

## Rolling
`.rolling()` provides rolling window, index must be datetime like

```python
df2_aapl = df_aapl.copy()
```

```python
df2_aapl = df2_aapl.drop(['year','month', 'day'], axis=1)
```

```python
# weekly rolling window, the first 6 days return NaN value because no data available
df2_aapl.rolling(7).mean().head(20)
```

```python
# 30 day rolling window
df2_aapl['Close 30 day mean'] = df2_aapl['Close'].rolling(window=30).mean()
```

```python
# plotting rolling 30 day Close on Close
df2_aapl[['Close','Close 30 day mean']].plot(figsize=(12,6))
```
