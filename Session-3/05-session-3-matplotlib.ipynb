# Matplotlib

Matplotlib is for Python plotting libraries. It's very powerfull and can create all sort of plots, giving a programmer complete control over the apperance of plots. Such control can sometimes however be a burden, since you have to specify many things to create what may be like a simple plot. There are two common methods for using Matplotlib `Figures and Axes` and `Pyplot`.

```python
# import necessary libraries
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
mpl.rcParams['figure.figsize'] = (13,8) # set default figure size 
```

```python
# if using older generation of jupyter notebooks add the below to display plot otherwise add plt.show()
# this cell only for older generation jupyter notebooks
%matplotlib inline
```

# Figures and Axes

A figure objects represent the whole image (or firgure) whereas and the axis represents a individual plot. A figure object may contain a single axis object or multiple.

```python
fig = plt.figure()  # nothing shows since its empty figure
```

```python
fig = plt.figure(figsize=(16,8)) # specify size in cm, width and height
ax = fig.add_subplot(111) # 1 row, 1 col
```

Modify the axis object to add labels, title, etc.

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.set(xlim=[0.5, 4.5],
       ylim=[-2, 8], 
       title='An Example Axes',
       ylabel='Y-Axis',
       xlabel='X-Axis');
```

The above code could also have been written `set_something` instead, this can get slightly tedious but it gives us more control, for example below I've bumped up the size of the title.

```python
fig = plt.figure(figsize=(13,8))
ax = fig.add_subplot(111) 
ax.set_xlim([0.5, 4.5])
ax.set_ylim([-2, 8])
ax.set_title('An Example Axes',size=25)
ax.set_ylabel('Y-Axis')
ax.set_xlabel('X-Axis')
```

Can create multiple axis in a single figure easily 

```python
fig, axes = plt.subplots(1,2) # rows, columns
```

```python
axes # axes is and array of axis objects
```

# Plotting

Up untill now we've only made axis object, lets draw something on them.


```python
x = np.linspace(1,10,100)
y = x ** 2
```

`.scatter` takes two arguremnts a x and and a y argument and draws a point.

```python
fig, ax = plt.subplots()
ax.scatter(x,y, label= 'x,y')
ax.scatter(x, y / 2, label= 'x,y / 2')
ax.legend()
```

Can also draw lines using `.plot` . If we only pass `plot` a single argument then the index will be used for the x-axis and the list's values for the y-axis.

```python
fig, axes = plt.subplots(1,2, figsize=(13,8))
axes[0].plot(x)
axes[1].plot(1/x, y)
```

## Axes methods vs. pyplot

Interestingly, just about all methods of `Figures and Axes` object exist as a function in the `pyplot` module (and vice-versa). For example, when calling `plt.xlim(1, 10)`, `pyplot` calls `ax.set_xlim(1, 10)` on whichever `Axes` is "current". Here is an equivalent version of the above example using just pyplot.

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
```

A single plot, will almost have all of the axes object functions on the pyplot module

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='x')
plt.xlim(0.5, 4.5);
```

# Markers
[Markers](http://matplotlib.org/api/markers_api.html) are commonly used in [`plot()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) and [`scatter()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter) plots, but also show up elsewhere. There is a wide set of markers available, and custom markers can even be specified.

marker     |  description  |marker    |  description    |marker    |  description  |marker    |  description  
----------|--------------|---------|----------------|---------|--------------|---------|--------------
"."        |  point        |"+"       |  plus           |","       |  pixel        |"x"       |  cross
"o"        |  circle       |"D"       |  diamond        |"d"       |  thin_diamond |          |
"8"        |  octagon      |"s"       |  square         |"p"       |  pentagon     |"\*"      |  star
"&#124;"   |  vertical line|"\_"      | horizontal line |"h"       |  hexagon1     |"H"       |  hexagon2
0          |  tickleft     |4         |  caretleft      |"<"       | triangle_left |"3"       |  tri_left
1          |  tickright    |5         |  caretright     |">"       | triangle_right|"4"       |  tri_right
2          |  tickup       |6         |  caretup        |"^"       | triangle_up   |"2"       |  tri_up
3          |  tickdown     |7         |  caretdown      |"v"       | triangle_down |"1"       |  tri_down
"None"     |  nothing      |`None`    |  nothing        |" "       |  nothing      |""        |  nothing

```python
xs, ys = np.mgrid[:4, 9:0:-1]
markers = [".", "+", ",", "x", "o", "D", "d", "", "8", "s", "p", "*", "|", "_", "h", "H", 0, 4, "<", "3",
           1, 5, ">", "4", 2, 6, "^", "2", 3, 7, "v", "1", "None", None, " ", ""]
descripts = ["point", "plus", "pixel", "cross", "circle", "diamond", "thin diamond", "",
             "octagon", "square", "pentagon", "star", "vertical bar", "horizontal bar", "hexagon 1", "hexagon 2",
             "tick left", "caret left", "triangle left", "tri left", "tick right", "caret right", "triangle right", "tri right",
             "tick up", "caret up", "triangle up", "tri up", "tick down", "caret down", "triangle down", "tri down",
             "Nothing", "Nothing", "Nothing", "Nothing"]
fig, ax = plt.subplots(1, 1, figsize=(7.5, 4))
for x, y, m, d in zip(xs.T.flat, ys.T.flat, markers, descripts):
    ax.scatter(x, y, marker=m, s=100)
    ax.text(x + 0.1, y - 0.1, d, size=14)
ax.set_axis_off()
plt.show()
```

## Linestyles
Line styles are about as commonly used as colors. There are a few predefined linestyles available to use. 

linestyle          | description
-------------------|------------------------------
'-'                | solid
'--'               | dashed
'-.'               | dashdot
':'                | dotted


```python
t = np.arange(0.0, 5.0, 0.2)
plt.figure(figsize=(13,8))
plt.plot(t, t, '-', t, t**2, '--', t, t**3, '-.', t, -t, ':');
```

Can use the above line styles on other charts too

```python
fig, ax = plt.subplots(1, 1)
ax.bar([1, 2, 3, 4], [10, 20, 15, 13], ls='-', ec='r', lw=5)
```

# Colours 

The easiest way to plot colours is to give each label a unique integer.


```python
from seaborn import load_dataset
```

```python
df= load_dataset('iris')
X = df.iloc[:,[1,2]].values

#convert strings to numbers
y = df.species.values
color_ints = df.species.factorize()[0]
```

```python
df.species.factorize()[0]
```

Can then use the `c` parameter to pass the colour label for  each point.

```python
plt.scatter(X[:,0],X[:,1], c=color_ints)
```

Can define colour by name or RGB hex codes also

```python
x = np.random.randint(1,20,10)

fig, ax = plt.subplots()
ax.plot(x, x+5, color='blue', label='blue')
ax.plot(x, x-5, color='#00FF00', label='lime')
ax.plot(x, x/5, color='orange', label='orange')
ax.plot(x, x*5, color='#FF00FF', label='magenta')
ax.legend()
ax.set_title('colours')
```
