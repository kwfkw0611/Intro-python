# Seaborn

Seaborn is great for the data exploration stage as we can make complex plots very easily and they also look great. This notebook contains various different plot types for you reference.

Seaborn is based on Matplotlib and is used for more statistical informative plots

[seaborn gallery](https://seaborn.pydata.org/examples/index.html)

[seaborn tutorial](https://seaborn.pydata.org/tutorial.html)

[seaborn api](https://seaborn.pydata.org/api.html)


```python
# import the necessary libraries
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# run this cell if on old generation of jupyter lab
%matplotlib inline
```

### Dataset
Work with datasets packaged with Seaborn

```python
df = sns.load_dataset('diamonds')
df.head()
```

```python
df.info()
```

### heatmap


```python
# correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap='plasma')
```

### pairplot
plot numerical data pairwise relationship in a dataset, can be used to quickly plot for an overview of correlation, histograms are plotted on the diagonal and scatter plots off the diagonal. 

```python
sns.pairplot(df)
```

### distribution plot

```python
sns.distplot(df['price'])
```

### violin plot
A combination of distribution and a boxplot

```python
sns.violinplot(x=df['cut'],y=df['price'])
plt.title('violin plot of price')
```

### scatter

```python
sns.scatterplot(df['carat'], df['price'], hue=df['cut'])
```

### bar

```python
sns.barplot(x= df['cut'].unique(),y=df.groupby('cut')['price'].mean())
plt.title('average price per cut')
```

### regression plot
plot data and linear regression plot

```python
sns.regplot(x='price', y='carat', data=df)
```

### count plot

```python
sns.countplot(df['clarity'],hue=df['color'])
```

### catplot
Plotting categorical data

```python
sns.catplot(x='cut', y='price', data=df, hue='color', col='clarity', col_wrap=2)
```
